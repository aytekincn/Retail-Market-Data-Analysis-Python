## Average of sales by sub category?

```python
retailMarket.groupby("Sub_Category").mean("Sales")
```

# Maximum 5 profit information?

```python
top_5_profits = retailMarket.nlargest(5, "Profit")
top_5_profits
```

# The sum of the sales figures of each product?

```python
retailMarket.groupby('Sub_Category')["Sales"].sum()
```

# Data visualization


```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
retailMarket.plot(x="State", y=["Sales"])
retailMarket.plot()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=retailMarket)
plt.xticks(rotation=45)
plt.title('Category Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sub_Category', y='Sales', data=retailMarket)
plt.title('Sub Category Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='State', y='Profit', data=retailMarket)
plt.xticks(rotation=45)
plt.title('Profits by State')
plt.tight_layout()
plt.show()
```
